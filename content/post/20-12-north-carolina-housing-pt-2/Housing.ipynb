```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("NC_Housing_Prices_2018.csv")
```

```python
df
```

```python
df[df["total_bedrooms"]!=df["total_rooms"]]
```

```python
df = df[df["population"]>0]
```

```python
df.plot(kind="scatter", x="longitude", y="latitude", alpha=0.4,
    s=df["population"]/100, label="Median House Value", figsize=(18,6),
    c="median_house_value", cmap=plt.get_cmap("jet"), colorbar=True,
)
plt.legend()
```

```python
sns.pairplot(df, vars=['population', 'households', 'median_income', 'median_house_value', 'total_bedrooms', 'total_rooms', 'housing_median_age'],
            diag_kind='kde')
```
